# Univariate Analysis

```python
# Histogram of Age
sns.histplot(df["age"], kde = True, bins = 10)
```

```python
#  Countplot of Gender
sns.countplot(x  = 'gender', data =  df)
```

```python
# Countplot of school type
sns.countplot(df['school_type'])
```

# Bivariate Analysis

```python
# Lineplot of study hours
sns.lineplot(x = 'study_hours', y= 'overall_score', data = df)
```

```python
df['final_grade'].unique()
```

```python
sns.barplot(x = 'study_hours', y= 'final_grade', data = df, color = 'blue')
```

# From the above barplot we can see that more the study hours, better the grade

```python
sns.barplot(x = 'study_method', y= 'final_grade', data = df, color = 'pink')
```

# From the above plot it can be interpredeted that the source of study didnt impact the final_grade


```python
# Barplot of comaprision of public and private schools and overall marks
sns.barplot(x = 'school_type', y = 'overall_score', data = df)
```

```python
sns.barplot(x = 'extra_activities', y= 'overall_score', data = df, color = 'blue')
```

```python
sns.barplot(x = 'parent_education', y= 'overall_score', data = df, color = 'red')
```

```python
# Scatterplot of attendance percentage and overall score
sns.scatterplot(x = 'attendance_percentage', y = 'overall_score', data = df)
```

# From the data set it can be concluded that, the student performance is mostly dependent upon study hours. Other factors like, study_mode or parent education, travel time do not impact the overall scores much

# Bivariate Analysis

```python
sns.barplot(x = 'gender', y = 'category', data = df)
```

```python
sns.barplot(x = 'price', y = 'category', data = df)
```

# From the above barplot it can be seen that, the price of technology, followed by shoes and clothing are the highest

```python
sns.barplot(x = 'payment_method', y = 'category', data = df)
```

```python
sns.barplot(x = 'shopping_mall', y = 'category', data = df)
plt.xticks(rotation = 45)
```

# From the above data analysis it is seen that the number or count of clothes is the highest, and the price of technology and clothes is the highest.
